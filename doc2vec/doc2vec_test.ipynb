```python
from gensim.test.utils import common_texts
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import gensim.downloader as api
import gensim
```

```python
def lower_case(text):
    return text.lower()

PUNC = '''!()-[]{};:'"\, <>./?@#$^&*~–“”|'''
def remove_punc(sentence):
    """Remove functuation in sentence"""
    in_sentence = sentence
    print(in_sentence)
    list_char = list(set(sentence))
    for char in list_char:
        if char in PUNC:
            sentence = sentence.replace(char, ' ')

    sentence = ' '.join(sentence.split())
    return sentence
def preprocess_text(text):
    text = lower_case(text)
    text = remove_punc(text)
    return text 
```

```python
dataset = api.load("text8")
data = [d for d in dataset] # list of list of words
```

```python
def tagged_document(list_of_list_of_words):
    for i, list_of_words in enumerate(list_of_list_of_words):
        yield gensim.models.doc2vec.TaggedDocument(list_of_words, [i])
data_for_training = list(tagged_document(data))
```

```python
from gensim.models.callbacks import CallbackAny2Vec

class callback(CallbackAny2Vec):
    
    def __init__(self):
        self.epoch = 0
    
    def on_each_epoch(self, model):
        loss = model.get_latest_training_loss()
        print('Loss after epoch {}: {}'.format(self.epoch, loss))
        self.epoch += 1
```

```python
model = gensim.models.doc2vec.Doc2Vec(vector_size=300, min_count=2, epochs=50, compute_loss=True, callbacks=[callback()])
```

```python
model.build_vocab(data_for_training)
```

```python
with open('test.txt', 'r', encoding='utf8') as file:
    text = file.read()

print(text)
```

```python
from pyvi.ViTokenizer import tokenize

text = tokenize(text)
text = preprocess_text(text)
print(text)
```

```python
score = model.docvecs.most_similar([model.infer_vector(text.split())], topn = 5)
```

```python
for id, scr in score:
    print("ID of the paragraph:", id, "The confidence score:", scr)
```
